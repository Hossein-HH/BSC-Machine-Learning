```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')
from sklearn.preprocessing import StandardScaler

#? Normalize the input data
A = np.array([[10, 8], [1, 2], [7, 5], [3, 5], [7, 6], [8, 7], [9, 9], [4, 5], [6, 5], [6, 8],
              [1, 9], [10, 2], [6, 3], [2, 5], [1, 14], [8, 8], [9, 5], [4, 4], [5, 6], [8, 8],
              [11, 9], [10, 12], [6, 4], [5, 2], [10, 2], [8, 3], [6, 9], [0, 4], [13, 6], [9, 6]])

A = StandardScaler(with_std=False, copy=False).fit_transform(A)
fig = plt.figure(figsize=(15, 10))
ax0 = fig.add_subplot(111)

#? we should equalization x , y Axis
ax0.set_aspect('equal')

#? moshakhas kardane mahdodeye mehvare x , y
ax0.set_ylim(bottom=min(A[:, 1]) - 3, top=max(A[:, 1]) + 3)

#? scatter yani noghat ro parakande rasm kon nachasbooneshon be ham dige
ax0.scatter(A[:, 0], A[:, 1])

#? initialize a first vector
v = np.array([1, 0.5])

ax0.plot(np.linspace(min(A[:, 0]) - 3, max(A[:, 0]) + 3),
         np.linspace(min(A[:, 0]) - 3, max(A[:, 0]) + 3) * (v[1] / v[0]),
         'k--', linewidth=1.5, zorder=0)

#? save projected points
coordinates_on_ba_run = []

for i in range(len(A[:, 0])):

    w = np.array([A[i][0], A[i][1]])

    #? zaribe K inja neveshtim c
    c = np.dot(w, v.reshape(2, 1)) / np.dot(v, v.reshape(2, 1))
    print(np.dot((w - c * v), v))

    #? cv is that result of projection we need to that
    cv = c * v

    #? draw the projected points
    ax0.plot([w[0], cv[0]], [w[1], cv[1]], linewidth=1, color="red", linestyle="--", zorder=0)

plt.show()

#!!!!!!!!!!!!!!!!!!!! _ERRORS!!!!!!!!
```

```python
from skspatial.objects import Vector
from skspatial.plotting import plot_2d

vector_a = Vector([1, 1])
vector_b = Vector([2, 0])

vector_projected = vector_b.project_vector(vector_a)

_, ax = plot_2d(
    vector_a.plotter(color='k', head_width=0.1),
    vector_b.plotter(color='k', head_width=0.1),
    vector_projected.plotter(color='r', head_width=0.1),
)

ax.axis([-0.5, 2.5, -0.5, 1.5])
```
